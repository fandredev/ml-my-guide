## Introdução ao Seaborn e Visualização de Dados
Este notebook demonstra como usar o Seaborn para visualização de dados, com foco nos dados de imigração para o Canadá.

```python
# Importando as bibliotecas necessárias para manipulação e visualização de dados
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o conjunto de dados e exibindo as primeiras linhas
df = pd.read_csv('../data/canadian_immegration_data.csv')

df.head()
```

```python
# Configurando o tema padrão para as visualizações do Seaborn
sns.set_theme()
```

```python
# Preparando os dados: definindo o índice e selecionando os 10 principais países por total de imigração
df.set_index('Country', inplace=True)
top_10 = df.sort_values(by='Total', ascending=False).head(10)

top_10
```

```python
# Criando um gráfico de barras horizontal para visualizar os 10 principais países por total de imigração
sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h', hue=top_10.index, legend=False)
plt.show()
```

```python
# Personalizando o gráfico de barras com títulos e rótulos
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set(title='Países com maior imigração para o Canadá\n(1980 a 2013)', xlabel='Número de imigrantes', ylabel='')

plt.show()
```

```python
# Ajustando o tamanho da figura e adicionando um título e rótulos mais detalhados
fig, ax = plt.subplots(figsize=(8, 4))

ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Países com maior imigração para o Canadá\n(1980 a 2013)', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=11, labelpad=30)
ax.set_ylabel('')

plt.show()
```

```python
# Definindo uma função para gerar gráficos de barras com diferentes paletas de cores
def generate_graphic_palette(palette):
  _, ax = plt.subplots(figsize=(8, 4))

  ax = sns.barplot(
        data=top_10,
        x='Total',
        y='Country',
        hue='Country',
        orient='h',
        palette=palette,
        legend=False
    )
  ax.set_title('Países com maior imigração para o Canadá \n(1980 a 2013)', loc='left', fontsize=18)
  ax.set_xlabel('Número de imigrantes', fontsize=14, labelpad=15)
  ax.set_ylabel('')
  plt.show()
```

```python
generate_graphic_palette('Blues_r')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

sns.barplot(data=top_10,
        x='Total',
        y='Country',
        hue='Country',
        orient='h',
        palette='tab10',
        legend=False)

ax.set_title('Países com maior imigração para o Canadá', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')

sns.despine()
plt.show()
```
