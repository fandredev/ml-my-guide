```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### 1. Write a Python programming to display a bar chart of the popularity of programming Languages.

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

df = pd.DataFrame({
    'Linguagem': languages,
    'Popularidade (%)': popularity
})

df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))

x = df['Linguagem']
y = df['Popularidade (%)']

ax.bar(x, y, color='blue', edgecolor='black')
ax.set_title('Popularity of Programming Language \nWorldwide, Oct 2017 compared to a year ago', fontsize=14, pad=10)

ax.set_xlabel('Languages', labelpad=20)
ax.set_ylabel('Popularity (%)', labelpad=20)
ax.set_ylim(0, 30)

plt.grid(color='r', linestyle='--', linewidth=0.5)
```

#### 4. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

df = pd.DataFrame({
    'Linguagem': languages,
    'Popularidade (%)': popularity
})

df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
x = df['Linguagem']
y = df['Popularidade (%)']

bar_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown']

for i in range(len(df)):
    ax.bar(x[i], y[i], color=bar_colors[i], edgecolor='black', label=x[i])

ax.set_title('Popularity of Programming Language \nWorldwide, Oct 2017 compared to a year ago', fontsize=14, pad=10)
ax.set_xlabel('Languages', labelpad=20, fontfamily='arial', fontsize=12, fontweight='bold')
ax.set_ylabel('Popularity (%)', labelpad=20, fontfamily='cursive', fontsize=12)
ax.set_ylim(0, 25)

handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), title="Languages", loc="upper right", fontsize=10, title_fontsize=12)

plt.grid(color='r', linestyle='--', linewidth=0.5)
plt.show()


fig.savefig('popularity_languages.png', dpi=300, bbox_inches='tight')
```

#### 5. Write a Python programming to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity (float value).

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

df = pd.DataFrame({
  'languages': languages,
  'popularity': popularity
})

x = df['languages']
y = df['popularity']

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))

fig.suptitle('Popularity of Programming Language\nWorldwide, Oct 2017 compared to a year ago')

bars = ax.bar(x, y, color='darkblue')

ax.bar_label(bars, padding=3)
ax.set_xlabel('Languages', labelpad=30)
ax.set_ylabel('Popularity', labelpad=30)

ax.grid(color='r', lw=0.33)

plt.legend(["Languages"], fontsize="large")

ax.set_yticks([0, 5, 10, 15, 20, 25])


plt.show()
```

#### 10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

```python
men = (22, 30, 35, 35, 26)
women = (25, 32, 30, 35, 29)
scores = (0, 10, 20, 30, 40)

df = pd.DataFrame({
  'men': men,
  'women': women
})
df
```

```python

plt.style.use('seaborn-v0_8-talk')
x = np.arange(len(scores))
bar_width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

bars1 = ax.bar(x - bar_width/2, df['men'], width=bar_width, label='Men', color='g', ec='black')
bars2 = ax.bar(x + bar_width/2, df['women'], width=bar_width, label='Women', color='crimson', ec='black')

ax.set_title('Scores by person', pad=19)
ax.set_xlabel('Person', labelpad=19)
ax.set_ylabel('Scores', labelpad=19)
ax.set_xticks(x)
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5'])

ax.legend()
ax.grid(False)

plt.show()
```
