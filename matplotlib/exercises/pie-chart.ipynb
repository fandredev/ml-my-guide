```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### 1. Write a Python programming to create a pie chart of the popularity of programming Languages.

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = np.array([22.2, 17.6, 8.8, 8.0, 7.7, 6.7])


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
explode = (0.1, 0, 0, 0, 0, 0)

ax.pie(popularity, labels=languages, autopct='%1.1f%%', pctdistance=0.7, shadow={
  'shade': 0.3,
  'ox': 0.01
}, startangle=140, explode=explode)

plt.show()
```

####  2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.

```python
# Criando um grfico de pizza para representar a popularidade de linguagens de programao
# Configurando o tamanho da figura
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))

# Definindo a exploso para destacar a primeira fatia (Java)
explode = (0.1, 0, 0, 0, 0, 0)

# Criando o grfico de pizza com as configuraes especificadas
ax.pie(
  popularity,  # Valores de popularidade
  labels=languages,  # Rtulos das fatias
  autopct='%1.1f%%',  # Formato das porcentagens
  pctdistance=0.75,  # Distncia das porcentagens em relao ao centro
  explode=explode,  # Exploso das fatias
  startangle=140,  # ngulo inicial
  radius=True,  # Ajuste do raio
  shadow={'ox': -0.01, 'edgecolor': 'none', 'shade': 0.3},  # Configurao de sombra
  wedgeprops={'edgecolor': 'black', 'linewidth': 0.5}  # Propriedades das bordas das fatias
)

# Adicionando um ttulo ao grfico
ax.set_title(
  'Popularidade de Linguagens de Programao \nMundialmente, Outubro de 2017 comparado ao ano anterior',
  pad=30,  # Espaamento entre o ttulo e o grfico
  fontweight='300',  # Peso da fonte
  bbox={'facecolor': '#C0C0C0'}  # Fundo do ttulo
)

# Ajustando a proporo do grfico
ax.set_aspect(0.88)

# Exibindo o grfico
plt.show()
```

#### 3. Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file.

```python
df = pd.read_csv('../../data/medal.csv')

df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

explode = (0.1,0, 0, 0, 0)

ax.pie(
  x=df['gold_medal'], 
  labels=df['country'],
  autopct='%1.1f%%', 
  startangle=130,
  shadow=True,
  wedgeprops={'edgecolor': 'black', 'linewidth': 0.5},
  explode=explode,
  radius=True
 )

ax.set_title(
  'Gold medal achievements of five most successful \ncountries in 2016 Summer Olympics'
)

ax.set_aspect(0.88)



plt.show()
```
