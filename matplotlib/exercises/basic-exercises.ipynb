### 1. Criar um gráfico de linha simples

- Objetivo: Plotar um gráfico de linha com os dados x = [1, 2, 3, 4, 5] e y = [2, 4, 6, 8, 10].

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])


fig = plt.figure(figsize=(15, 5))
plt.bar(x, y, label="Exercicio 1", color="g")

plt.ylabel('Eixo Y', labelpad=30)
plt.xlabel('Eixo X', labelpad=30)

plt.legend()
plt.title('Exercicio 1')

plt.show()
```

### 2. Adicionar título e rótulos aos eixos

- Objetivo: Pegue o gráfico anterior e adicione:

- Título: "Gráfico de Crescimento"

- Rótulo no eixo X: "Tempo"

- Rótulo no eixo Y: "Valor"

### 4. Criar um gráfico de barras

- Objetivo: Crie um gráfico de barras com os seguintes dados:

categorias = ['A', 'B', 'C', 'D']

valores = [23, 45, 56, 78]

```python
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]

fig = plt.figure(figsize=(10, 5))

plt.bar(categorias, valores, color="m", width=0.3, label="Categorias e valores")
plt.xlabel('Categorias', labelpad=30)
plt.ylabel('Valores', labelpad=30)

plt.title('Categorias e valores')
plt.legend()

plt.show()
```

### 5. Adicionar cores personalizadas ao gráfico de barras

- Objetivo: Pinte cada barra com uma cor diferente.

```python
x = np.array([3,6,9,15,18])
y = np.array([3,6,9,15,18])

colors_list = ['#FF5733', '#33FF57', '#3357FF', '#FF33F6', '#F6FF33']

plt.bar(x, y, color=colors_list, label="Multiple Colors")


plt.legend()
plt.title('Multiple Colors')

plt.xlabel('Eixo X', labelpad=30)
plt.ylabel('Eixo Y', labelpad=15)

plt.show()
```

### 6. Criar um gráfico de pizza (pie chart)

- Objetivo: Com os dados abaixo, crie um gráfico de pizza:

categorias = ['Python', 'Java', 'C++', 'JavaScript']

porcentagens = [30, 25, 15, 30]


```python
categorias = ['Python', 'Java', 'C++', 'JavaScript']
porcentagens = [30, 25, 15, 30]


fig = plt.figure(figsize=(13,9))


plt.pie(porcentagens, labels=categorias, autopct='%1.1f%%', pctdistance=0.7, startangle=90)

plt.legend(loc="upper left")

plt.show()
```
