#### 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
plt.style.use('tableau-colorblind10')
```

```python
x = np.arange(0, 60, 10)
y = np.arange(0, 60, 10)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 3))

fig.suptitle('Draw a line.')

ax.set_xlabel('x-axis', labelpad=10, fontsize=13, fontfamily='arial')
ax.set_ylabel('y-axis', labelpad=10, fontsize=13, fontfamily='cursive')

plt.plot(x, y)

plt.ylim(0, 170)
plt.show()
```

#### 2. Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis and a title.

#### 4. Write a Python program to plot two or more lines on same plot with suitable legends of each line.

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(7, 5))


# Dados de exemplo
x = [10, 20, 30]
y1 = [20, 40, 10]  # line 1
y2 = [40, 10, 30]  # line 2

# Criar figura
ax.plot(x, y1, color='blue', label='line 1')
ax.plot(x, y2, color='green', label='line 2')

# Adicionar título e rótulos
ax.set_title('Two or more lines on same plot with suitable legends')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')

# Adicionar legenda
plt.legend()

# Adicionar mudanças no eixo x
ax.set_xticks([10, 15, 20, 25, 30])

# Exibir o gráfico
plt.show()
```

#### 6. Write a Python program to plot two or more lines with legends, different widths and colors.

```python
plt.style.use('seaborn-v0_8-notebook')

fig, ax = plt.subplots(figsize=(7, 5))

# Dados de exemplo
x = [10, 20, 30]
y1 = [20, 40, 10]  # line 1
y2 = [40, 10, 30]  # line 2

# Criar figura
ax.plot(x, y1, color='blue', label='line1-width-3', lw=3)
ax.plot(x, y2, color='green', label='line2-width-5', lw=5)

# Adicionar título e rótulos
ax.set_title('Two or more lines on same plot with suitable legends', pad=20)
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')

# Adicionar legenda
plt.legend(loc='upper right', labelspacing=1.5, fontsize=10, shadow=True)

# Adicionar mudanças no eixo x
ax.set_xticks([10, 15, 20, 25, 30])

# Exibir o gráfico
plt.show()
```

#### 7. Write a Python program to plot two or more lines with different styles.

```python
x = [10, 20, 30]
l1 = [20, 40, 10]
l2 = [40, 10, 30]

fig, ax = plt.subplots(figsize=(5, 5))

fig.suptitle('Plot with two or more lines with different styles')

ax.plot(x, l1, ls='dotted', lw=1.7, color='darkblue', label='line1-dotted')
ax.plot(x, l2, ls='dashed', lw=3, label='line2-dashed')

ax.set_xlabel('x-axis')
ax.set_xlabel('x-axis')

plt.legend(loc='upper right', labelspacing=1.5, fontsize=10, shadow=True)

ax.set_xticks([10, 15, 20, 25, 30])

plt.show()
```

#### 8. Write a Python program to plot two or more lines and set the line markers.

```python
x = np.array([1, 4, 5, 6, 7])
y = np.array([2, 6, 3, 6, 3])

x, y
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(9, 5))

fig.suptitle('Display marker', fontweight='bold', fontsize=19)

# ms = markersize
ax.plot(x, y, marker='.', ms=19, ls='dotted', mfc='b')

ax.set_xlabel('x-axis', fontsize=13, labelpad=10, color='g')
ax.set_ylabel('y-axis', fontsize=13, labelpad=10, color='g')


ax.set_yticks(np.arange(1, 9, 1))
ax.set_xticks(np.arange(1, 9, 1))

plt.show()
```

#### 9. Write a Python program to display the current axis limits values and set new axis values.

```python
x = np.array([0, 50])
y = np.array([0, 100])

x, y
```

```python
# Set style

plt.style.use('fivethirtyeight')


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 5))

fig.suptitle('Draw a line')

ax.plot(x, y, lw=1.8, color='darkblue')

ax.set_xlabel('x-axis', labelpad=15)
ax.set_ylabel('y-axis', labelpad=15)

ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_yticks([0, 50, 100, 150, 200])

plt.show()
```

#### 10. Write a Python program to plot quantities which have an x and y position.

```python
plt.style.use('fast');

x = [2, 3, 5, 6, 8]
y = [0, 5, 10, 17, 20]

xx = [3, 4, 6, 7, 9]
yy = [2, 6, 11, 20, 22]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

ax.plot(x, y, marker='*', markersize=5, linewidth=0, color='darkblue')
ax.plot(xx, yy, marker='o', markersize=5, linewidth=0, color='r')

ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_yticks([0, 5, 10, 15, 20, 25, 30])

plt.show()
```
