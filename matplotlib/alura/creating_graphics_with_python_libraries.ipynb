```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../data/canadian_immegration_data.csv')

df.head()
```

```python
df.info()
```

```python

# Alterando o index
df.set_index('Country', inplace=True)
```

```python
years = list(map(str, range(1980,	2014)))
```

```python
years
```

```python
# Buscar apenas anos do Brasil

only_brazil_data = df.loc['Brazil', years]
only_brazil_data
```

```python
brazil_dict = {
  "years": only_brazil_data.index.tolist(),
  "immigrants": only_brazil_data.values.tolist()
}

data_brazil = pd.DataFrame(brazil_dict)
```

```python
data_brazil
```

```python
# Altera como a figura deve se comportar
plt.figure(figsize=(8, 4))

# Adicionando no Eixo Y os anos, e no eixo X os imigrantes
plt.plot(data_brazil['years'], data_brazil['immigrants'])

# Adicionando um tÃ­tulo para a imagem
plt.title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡')

# Adicionando uma label para o eixo X
plt.xlabel('Ano', labelpad=30)

# Adicionando uma label para o eixo Y
plt.ylabel('NÃºmero de imigrantes', labelpad=30)

# Altera o comportamento dos dados do eixo X [Horizontal]
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

# Mostra o grÃ¡fico e retira mensagens de log da library
plt.show()
```

```python
IPython_default = plt.rcParams.copy()

# Usa um tema jÃ¡ do matplotlib
with plt.style.context('fivethirtyeight'):
# Cria uma figura com 8x4 polegadas
  fig, ax = plt.subplots(figsize=(8, 4))

  # Plotando dados no grÃ¡fico dentro de uma figura
  ax.plot(data_brazil['years'], data_brazil['immigrants'], lw=1.8, color='g')

  # Adicionando um tÃ­tulo
  ax.set_title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡ (1980 a 2013)', fontsize=16, loc='left')

  # Adicionando rÃ³tulo para o eixo X
  ax.set_xlabel('Ano', labelpad=15, fontsize=13)

  # Adicionando rÃ³tulo para o eixo Y
  ax.set_ylabel('Imigrantes', labelpad=15, fontsize=13)

  # Altera as propriedades das labels do eixo x e y
  ax.xaxis.set_tick_params(labelsize=12)
  ax.yaxis.set_tick_params(labelsize=12)

  # Colocando uma grade no grÃ¡fico
  ax.grid(linestyle='--')

  # Modificando o eixo X dentro de uma imagem com 5 de frequÃªncia (1980, 1985, 1990..)
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

  # Salva o grÃ¡fico como uma imagem com nÃ£o transparencia no fundo da imagem
  fig.savefig('imigraÃ§Ã£o_brasil_to_canada.png', transparent=None, dpi=300, bbox_inches='tight')


  plt.show()
```

```python
south_america = df.query('Region == "South America"').sort_values(by='Total', ascending=True)
south_america
```

#### CriaÃ§Ã£o de um grafico de barras com mÃºltiplas cores e anotaÃ§Ãµes

```python
fig, ax = plt.subplots(figsize=(12, 5))

# colors = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
colors = []

for country in south_america.index:
  if country == 'Brazil':
    colors.append('green')
  else:
    colors.append('silver')

ax.barh(south_america.index, south_america['Total'], color=colors)
ax.set_title('AmÃ©rica do Sul: Brasil foi o quarto paÃ­s com mais imigrantes\npara o CanadÃ¡ no perÃ­odo (1980-2013)', loc='left', fontsize=14)

ax.set_xlabel('NÃºmero de imigrantes', labelpad=33)
ax.set_ylabel('', fontsize=14)

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Adicionando valores corretos ao lado de cada barra
for i, v in enumerate(south_america['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')


# Removendo frame do grÃ¡fico
ax.set_frame_on(False)

# ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='major', length=0)

fig.savefig('imigraÃ§Ã£o_americas_do_sul_to_canada.png', transparent=None, dpi=300, bbox_inches='tight')

plt.show()
```

#### CriaÃ§Ã£o de Subplots [Multiplos GrÃ¡ficos]

```python
# Plotando dois grÃ¡ficos na mesma imagem
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Manipulando o primeiro grÃ¡fico
axs[0].plot(data_brazil['years'], data_brazil['immigrants'])
axs[0].set_title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡\n(1980 a 2013)')
axs[0].set_xlabel('Ano', labelpad=15)
axs[0].set_ylabel('NÃºmero de imigrantes', labelpad=15)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[0].grid()


# Manipulando o segundo grÃ¡fico
axs[1].boxplot(data_brazil['immigrants'])
axs[1].set_title('Boxplot da imigraÃ§Ã£o do Brasil para o CanadÃ¡\n(1980 a 2013)')
axs[1].set_xlabel('Brasil', labelpad=15)
axs[1].set_ylabel('NÃºmero de imigrantes', labelpad=15)

axs[1].grid()


axs[0].spines['top'].set_visible(False) # Remove linha lateral no top
axs[0].spines['right'].set_visible(False) # Remove linha lateral a direita

plt.show()
```

```python
# CriaÃ§Ã£o de uma figura com 4 subplots (2x2) com tamanho 10x6 polegadas
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Ajusta o espaÃ§amento entre os subplots
# hspace: espaÃ§amento vertical entre os grÃ¡ficos
# wspace: espaÃ§amento horizontal entre os grÃ¡ficos
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Define um tÃ­tulo principal para toda a figura
fig.suptitle('ImigraÃ§Ã£o dos 4 maiores paÃ­ses da AmÃ©rica do Sul para o CanadÃ¡\n(1980 a 2013)')

# Cria o grÃ¡fico do Brasil no subplot superior esquerdo [0,0]
axs[0, 0].plot(years, df.loc['Brazil', years])
axs[0, 0].set_title('Brasil')

# Cria o grÃ¡fico da ColÃ´mbia no subplot superior direito [0,1]
axs[0, 1].plot(years, df.loc['Colombia', years])
axs[0, 1].set_title('ColÃ´mbia')

# Cria o grÃ¡fico da Argentina no subplot inferior esquerdo [1,0]
axs[1, 0].plot(years, df.loc['Argentina', years])
axs[1, 0].set_title('Argentina')

# Cria o grÃ¡fico do Peru no subplot inferior direito [1,1]
axs[1, 1].plot(years, df.loc['Peru', years])
axs[1, 1].set_title('Peru')

# Configura os eixos X e Y para todos os subplots
# Usando flat para iterar sobre todos os subplots como uma lista Ãºnica
for ax in axs.flat:
  # Define marcadores no eixo X a cada 5 anos
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  # Adiciona rÃ³tulos aos eixos com padding de 15
  ax.set_xlabel('Ano',labelpad=15)
  ax.set_ylabel('NÃºmero de imigrantes', labelpad=15)

# Define os limites do eixo Y
ymin = 0
ymax = 7000

# Aplica os mesmos limites do eixo Y para todos os subplots
# ravel() Ã© similar ao flat, transforma matriz em array 1D
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

# Exibe a figura com todos os subplots
plt.show()
```
