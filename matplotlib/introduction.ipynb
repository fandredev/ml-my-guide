```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)

plt.ylabel('Eixo Y') # Cria uma legenda para o eixo Y
plt.xlabel('Eixo X') # Cria uma legenda para o eixo X
plt.title('Título do Gráfico') # Cria um título para o gráfico
plt.grid() # Cria uma grade no gráfico
plt.xticks([0, 4, 6, 8, 9]) # Cria os pontos do eixo X
plt.yticks([0, 10, 20, 30]) # Cria os pontos do eixo Y

plt.legend(['Dados']) # Cria uma legenda para a linha

plt.show() # Mostra o gráfico
```

## Lines Properties

Algumas propriedades para personalizar seu gráfico de linha

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, linestyle=':', color='b', marker='o', markersize=5, label='Dados', lw=1.5) # Cria uma linha pontilhada azul com marcadores circulares de tamanho 5 e largura 1.5
plt.ylabel('Eixo Y') # Cria uma legenda para o eixo Y
plt.xlabel('Eixo X') # Cria uma legenda para o eixo X
plt.title('Título do Gráfico') # Cria um título para o gráfico
plt.grid() # Cria uma grade no gráfico
plt.show()
```

## Gráfico de Scatter

Gráficos de dispersão.

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]


plt.scatter(x, y, color="b", marker="o", s=50, label="Dados", lw=1) # Cria um gráfico de dispersão com marcadores azuis de tamanho 50
plt.ylabel('Eixo Y') # Cria uma legenda para o eixo Y
plt.xlabel('Eixo X') # Cria uma legenda para o eixo X
plt.title('Título do Gráfico') # Cria um título para o gráfico
plt.grid() # Cria uma grade no gráfico
plt.legend(['Dados do gráfico']) # Cria uma legenda para o gráfico
plt.show()
```

## Gráfico de barras

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.bar(x, y, color="r", label="Dados") # Cria um gráfico de barras vermelho com largura 1
plt.ylabel('Eixo Y') # Cria uma legenda para o eixo Y
plt.xlabel('Eixo X') # Cria uma legenda para o eixo X
plt.title('Título do Gráfico') # Cria um título para o gráfico
plt.legend(['Dados do gráfico']) # Cria uma legenda para o gráfico
plt.show()
```

## Alterando os Markers do Gráfico

- https://matplotlib.org/stable/api/markers_api.html


```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]


plt.scatter(x, y, color="r", marker="p") # Cria um gráfico de dispersão com marcadores vermelhos em forma de losango 
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico com marcadores em forma de losango')
plt.legend(['Dados do gráfico']) 
plt.show()
```

## Alterando limites do gráfico

```python
plt.plot(x, y, color="r", marker="p") # Cria um gráfico de dispersão com marcadores vermelhos em forma de losango
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico com marcadores em forma de losango')
plt.legend(['Dados do gráfico'])

plt.axis(xmin=1, xmax=10) # Define os limites do eixo X 
plt.axis(ymin=0, ymax=50) # Define os limites do eixo Y
plt.show()
```

## Criando multiplos gráficos usando uma imagem

```python
values_x = [1, 2, 3, 4]
values_y = [10, 20, 25, 30]

figure = plt.figure(figsize=(15,3)) # Cria uma figura com tamanho 10x5

figure.suptitle('Gráficos alinhados') # Cria um título para o gráfico

figure.add_subplot(1, 3, 1) # Cria um gráfico na posição 1
plt.plot(values_x, values_y, color="r", marker="p") # Cria um gráfico de dispersão com marcadores vermelhos em forma de losango
plt.legend(['Gráfico'])
plt.title('Gráfico 1') # Cria um título para o gráfico
plt.grid()

# Gráfico 2

figure.add_subplot(1, 3, 2) # Cria um gráfico na posição 2
plt.bar(values_x, values_y) # Cria um gráfico de dispersão com marcadores verdes em forma de círculo
plt.legend(['Gráfico de barras'])
plt.title('Gráfico 2') # Cria um título para o gráfico
plt.grid()

# Gráfico 3
figure.add_subplot(1, 3, 3) # Cria um gráfico na posição 3
plt.scatter(values_x, values_y, color="r", marker="p") # Cria um gráfico de dispersão com marcadores vermelhos em forma de losango
plt.legend(['Gráfico de dispersão'])
plt.title('Gráfico 3') # Cria um título para o gráfico
plt.grid()

plt.tight_layout() # Ajusta o layout dos gráficos

plt.savefig('graficos_alinhados.png') # Salva a figura com o nome 'graficos_alinhados.png'

plt.show(figure)
```

## Caso Real

```python
import pandas as pd

cotation_df = pd.read_csv('../data/indexData.csv')

display(cotation_df.head()) # Mostra as 5 primeiras linhas do dataframe
```

```python
cotation_df1 = cotation_df[['Index', 'Date','Close']] # Cria um novo dataframe com as colunas 'Index', 'Date' e 'Close'
```

```python
df_nya = cotation_df1.query('Index == "NYA"')

df_nya = df_nya[13940:] # Pegando os dados a partir do índice 13940

df_nya
```

```python

plt.figure(figsize=(15,3)) # Cria uma figura com tamanho 15x3
plt.plot(df_nya['Date'], df_nya['Close']) # Cria um gráfico de dispersão com marcadores vermelhos em forma de losango
plt.ylabel('Valor de fechamento') # Cria uma legenda para o eixo Y
plt.xlabel('Data da cotação')

plt.legend(['NYA'], fontsize=9) # Cria uma legenda para o gráfico
plt.title('Gráfico de cotação histórica da bolsa de NYA') # Cria um título para o gráfico

plt.grid() # Cria uma grade no gráfico
plt.xticks(rotation=45) # Rotaciona os rótulos do eixo X em 45 graus

plt.savefig('grafico_cotacao.png') # Salva a figura com o nome 'grafico_cotacao.png'

plt.show()
```
