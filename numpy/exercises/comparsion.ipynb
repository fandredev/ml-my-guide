##### 🧪 Exercício 3: Filtro com múltiplas condições

Crie um array de 20 números inteiros entre 1 e 100. Filtre os elementos que são pares e maiores que 50.

🔎 Dica: Combine condições com & e use () ao redor de cada uma. Exemplo: (arr % 2 == 0) & (arr > 50)
👉 Boolean indexing – Docs

```python
np.random.seed(4)

arr = np.random.randint(101, size=100)
arr
```

```python
condition = (arr > 50) & (arr % 2 == 0)
filtered_arr = arr[condition]
filtered_arr
```

```python
np.sort(filtered_arr)
```

```python
len(filtered_arr)
```

```python
np.unique(filtered_arr)
```

##### 🧪 Exercício 4: Comparando médias

Crie dois arrays com 10 valores aleatórios entre 0 e 1. Verifique qual array tem a média maior.

🔎 Dica: Use np.mean() para calcular médias e > para comparar.
👉 np.mean – Docs

```python
arr = np.random.rand(10)
arr2 = np.random.rand(10)

media_arr1 = np.mean(arr)
media_arr2 = np.mean(arr2)

greater_median = media_arr1 > media_arr2

msg = f'A média 1 é {"maior" if greater_median else "menor"} que a média dois.\nMédia N1: {media_arr1} \nMédia N2: {media_arr2}'
print(msg)
```

##### 🧪 Exercício 5: Quantos valores estão dentro de um intervalo

Dado um array de 15 valores entre 0 e 100, conte quantos valores estão entre 30 e 60, inclusive.

🔎 Dica: Combine condições com &, e depois use np.sum() para contar True.

```python
np.random.seed(15)

arr = np.random.randint(100, size=15)

arr
```

```python
condition = (arr >= 30) & (arr <= 60)
sum_between = np.sum(condition)
values_between = arr[condition]

print(f"Number of values between 30 and 60: {sum_between}")
print("Values between 30 and 60:", values_between)
```
