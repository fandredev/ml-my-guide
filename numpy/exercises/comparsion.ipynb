##### ğŸ§ª ExercÃ­cio 5: Quantos valores estÃ£o dentro de um intervalo

Dado um array de 15 valores entre 0 e 100, conte quantos valores estÃ£o entre 30 e 60, inclusive.

ğŸ” Dica: Combine condiÃ§Ãµes com &, e depois use np.sum() para contar True.

```python
np.random.seed(15)

arr = np.random.randint(100, size=15)

arr
```

```python
condition = (arr >= 30) & (arr <= 60)
sum_between = np.sum(condition)
values_between = arr[condition]

print(f"Number of values between 30 and 60: {sum_between}")
print("Values between 30 and 60:", values_between)
```
