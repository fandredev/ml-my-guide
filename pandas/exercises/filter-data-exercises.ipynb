````python
import pandas as pd

dados = {
    'nome': ['Alice', 'Bob', 'Carla', 'Daniel', 'Eduarda', 'Felipe'],
    'idade': [25, 35, 45, 28, 32, 40],
    'salario': [4000, 6000, 7000, 3000, 5500, 8000],
    'departamento': ['TI', 'Financeiro', 'TI', 'RH', 'TI', 'Marketing'],
    'cidade': ['São Paulo', 'Rio de Janeiro', 'São Paulo', 'Belo Horizonte', 'Curitiba', 'Porto Alegre'],
    'populacao': [12345678, 6747815, 12345678, 2512070, 1966553, 1484941],
    'data_nascimento': ['1997-05-15', None, '1978-11-30', '1993-07-21', '1990-03-10', '1985-12-05'],
    'categoria': ['Alimentos', 'Bebidas', 'Alimentos', 'Eletrônicos', 'Bebidas', 'Alimentos'],
    'status': ['ativo', 'inativo', 'ativo', 'ativo', 'inativo', 'ativo'],
    'valor': [10.5, 20.8, 15.0, 7.3, 18.6, 12.4],
    'nota': [8.5, 6.2, 9.0, 7.0, 8.0, 7.8],
    'aprovado': [True, False, True, True, False, True]
}

df = pd.DataFrame(dados)

````

### 1. Filtro Simples
##### Dado um DataFrame denominado df com uma coluna idade, crie um filtro para selecionar somente as linhas onde a idade seja maior que 30. Elabore o código sem executar a operação.

````python
# Filtra os dados onde a idade é maior que 30

df.query('idade > 30')
````

### 2. Filtro com Condição Múltipla (AND)
##### Utilizando um DataFrame chamado df que possui as colunas salario e departamento, crie um filtro para selecionar os funcionários que pertencem ao departamento “TI” e possuem salário superior a 5000. Demonstre a construção deste filtro utilizando operadores lógicos apropriados.

````python
# Filtra os funcionários que trabalham no departamento de TI e têm salário maior que 5000

df.query('departamento == "TI" and salario > 5000')
````

### 3. Filtro com Condição Múltipla (OR)
##### Considere um DataFrame df contendo as colunas cidade e populacao. Construa um filtro que retorne as linhas onde a cidade seja “São Paulo” ou a população seja maior que 1 milhão. Elabore a linha de código que realize essa filtragem.

````python
# Filtra os dados para mostrar apenas os funcionários que trabalham em São Paulo ou têm mais de 1 milhão de habitantes

minimum_population = 1_000_000

df.query('(cidade == "São Paulo") or (populacao > @minimum_population)')
````

### 4. Filtragem com Valores Nulos e Não Nulos
##### *Suponha que o DataFrame df possua uma coluna chamada data_nascimento, com alguns valores nulos. Crie exercícios para:

- Selecionar as linhas em que data_nascimento é nula.

- Selecionar as linhas onde data_nascimento não é nula.*

````python
# Seleciona só linhas que não são nulas

df[df['data_nascimento'].notnull()]
````

````python
# Seleciona só linhas que são nulas

df[~df['data_nascimento'].notnull()]
````

### 5. Utilização do Método .loc para Filtragem

##### A partir de um DataFrame df que possui as colunas produto, preco e categoria, crie um exercício para filtrar todas as linhas com categoria igual a “Eletrônicos” e retornar apenas as colunas produto e preco utilizando o método .loc.

````python
# Filtra o DataFrame para mostrar apenas as linhas onde a categoria é "Eletrônicos" e seleciona as colunas 'categoria' e 'valor'

df.loc[df['categoria'] == "Eletrônicos", ['categoria', 'valor']]
````

### 6. Filtragem Utilizando o Método ```query``` 

##### *Utilize o método query em um DataFrame df que possui as colunas nota e aprovado (booleano) para:

- Selecionar as linhas onde a nota seja maior ou igual a 7 e o aluno esteja aprovado.

````python
# Filtra notas maiores ou iguais a 7 e aprovados

df.query('nota >= 7 and aprovado')
````

### 7. Filtro com Lista de Valores ```(isin)```

##### *Considere um DataFrame df com uma coluna categoria que possui diversos valores categóricos. Desenvolva um exercício onde é necessário filtrar apenas as linhas cujas categorias estão presentes em uma lista, por exemplo, ['Alimentos', 'Bebidas']. Construa a linha de código utilizando o método isin.

````python
# Seleciona linhas onde a categoria é Alimentos ou Bebidas

df[df['categoria'].isin(['Alimentos', 'Bebidas'])]
````

### 8. Filtragem com Operador de Negação

##### A partir de um DataFrame df com uma coluna status, onde os possíveis valores são “ativo” e “inativo”, crie um exercício para filtrar as linhas onde o status não seja “inativo”. Explore o uso do operador ~ para inverter a condição.

````python
# Seleciona linhas onde o status é diferente de "inativo"

df.query('~(status == "inativo")')
````

### 9. Criação de Filtros Dinâmicos com Variáveis

##### Suponha que as variáveis ```min_valor``` e ```max_valor``` definam um intervalo para a coluna valor em um DataFrame df. Elabore um exercício para filtrar todas as linhas cujo valor esteja compreendido entre ```min_valor``` e ```max_valor```. O desafio é construir o filtro de forma dinâmica, usando as variáveis definidas.

````python
# Filtrando por intervalo de valores

min_valor = 7
max_valor = 13

df.query('valor >= @min_valor and valor <= @max_valor')
````
