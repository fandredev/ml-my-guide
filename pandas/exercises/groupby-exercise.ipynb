#### 1. Total de produtos vendidos por funcionário

Pergunta:
Use o groupby para descobrir quantos produtos cada funcionário vendeu no total (sum da coluna quantity).

```python
df = pd.read_csv('../../data/sales.csv')

df
```

```python
df.groupby('employee').sum()
```

#### 2. Valor total de vendas por departamento
Pergunta:
Calcule o valor total vendido (quantity * price) por departamento.

Dica: Você pode primeiro criar uma nova coluna chamada total_sale e depois agrupar.


```python
df['total_sale'] = df['quantity'] * df['price']

total_sales_by_department = df.groupby('department')['total_sale'].sum()
total_sales_by_department
```

#### 3. Quantidade média de produtos vendidos por produto


Pergunta:
Descubra a quantidade média vendida para cada produto (mean da coluna quantity agrupada pelo product).

```python
df

df.groupby('product').mean('quantity')
```

#### 4. Maior venda individual por funcionário

Pergunta:
Qual foi a maior quantidade de um único produto vendida por cada funcionário? (usando max da quantity agrupado por employee)

```python
df.groupby('employee').max('quantity').value_counts(normalize=True).to_frame()
```

### 5. Quantas vendas cada departamento realizou?
