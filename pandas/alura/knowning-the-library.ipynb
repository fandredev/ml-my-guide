# Manipulando os dados

## Criando colunas numéricas

```python
# Importando dados

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'

new_df = pd.read_csv(url, sep=";")

```

```python
# Criando uma coluna

new_df['Valor_por_mes'] = new_df['Valor'] + new_df['Condominio']
new_df.head()
```

```python
# Criando uma nova coluna

new_df['Valor_por_ano'] = (new_df['Valor'] * 12) + new_df['IPTU']

new_df.head()
```

## Criando colunas categóricas

```python
# Criando colunas novamente

new_df['Descricao'] = new_df['Tipo'] + ' em ' + new_df['Bairro']

new_df.head()
```

```python
# Erro: Não posso concatenar colunas strings com inteiras

new_df['Descricao'] = new_df['Tipo'] + ' em ' + new_df['Bairro'] + ' com ' + \
                                        new_df['Quartos'] + ' quarto(s) ' + \
                                        ' e ' + new_df['Vagas'] + ' vaga(s) de garagem.'
```

```python
# Resolvendo problema de int e srt p concatenar

new_df['Descricao'] = new_df['Tipo'] + ' em ' + new_df['Bairro'] + ' com ' + \
                                        new_df['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + new_df['Vagas'].astype(str) + ' vaga(s) de garagem.'
```

```python
new_df
```

```python
# Criando uma coluna nova a partir de uma coluna existente e aplicando uma lógica usando lambda functions

new_df['Possui_suite'] = new_df['Suites'].apply(lambda s: 'Sim' if s > 0 else 'Não')

new_df
```

```python
# Finalizando salvando o arquivo final

new_df.to_csv('finish_data_dev.csv', index=False, sep=";")
```
