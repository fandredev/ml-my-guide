### Trabalhando com CSV

```python
import pandas as pd
```

```python
# Lendo o arquivo com o separador padrão

url = 'https://raw.githubusercontent.com/alura-cursos/pandas/refs/heads/main/superstore_data.csv'
data_market = pd.read_csv(url)
```

```python
data_market.head()
```

```python
# Lendo o arquivo com o separador padrão pelo separador de ponto e vírgula

url_two = 'https://raw.githubusercontent.com/alura-cursos/pandas/refs/heads/main/superstore_data_ponto_virgula.csv'

data_market_point_comma = pd.read_csv(url_two, sep=";")

data_market_point_comma.head()
```

### Trabalhando com arquivos JSON

```python
patients_json = pd.read_json('https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/pacientes.json')
```

```python
patients_json.head()
```

```python
patients_data_2 = pd.read_json('https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/pacientes_2.json')

patients_data_2
```

```python
# Transforma JSON aninhados em colunas e linhas no Dataframe

df_normalize = pd.json_normalize(patients_data_2['Pacientes'])
```

```python
df_normalize
```

```python
# Salvando em .json

df_normalize.to_json('historic_patients_normalized.json')
```

```python
pd.read_json('historic_patients_normalized.json')
```
