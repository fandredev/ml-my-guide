### Trabalhando com SQL

```python
from sqlalchemy import create_engine, inspect, text
```

```python
# Cria um banco de dados sqlite em memória

engine = create_engine('sqlite:///:memory:')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv')
```

```python
df.head()
```

```python
# Cria uma tabela chamada clientes para x engine baseada no DataFrame

df.to_sql('clients', engine, index=False)
```

```python
# Inspeciona o banco de dados

inspector_db = inspect(engine)
```

```python
# Busca nomes das tabelas

print(
  inspector_db.get_table_names()
)
```

```python
query = "SELECT * FROM clients WHERE Categoria_de_renda = 'Empregado'"
```

```python
employees = pd.read_sql(query, engine)
```

```python
employees
```

```python
# Criando uma nova tabela

employees.to_sql('employees', con=engine, index=False)
```

```python
# Ler tabela especifica


pd.read_sql_table('employees', engine)
```

```python
# Apenas algumas colunas

pd.read_sql_table('employees', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
query = 'SELECT * FROM clients'
```

```python
pd.read_sql(query, engine)
```

```python
query = 'DELETE FROM clients WHERE ID_Cliente = 5008804'
```

```python
# Executa a query e commita dentro do banco de dados

with engine.connect() as conn:
  conn.execute(text(query))
  conn.commit()
```

```python
pd.read_sql_table('clients', con=engine)
```

```python
query = 'UPDATE clients SET Grau_escolaridade = "Ensino Superior" where ID_Cliente = 5008808'
```

```python
with engine.connect() as conn:
  conn.execute(text(query))
  conn.commit()
```

```python
pd.read_sql_table('clients', con=engine)
```
