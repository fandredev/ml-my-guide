```python
import pandas as pd
import numpy as np
```

```python
# Lendo esse JSON, o Dataframe ficará diferente pois ficarao como objetos dentro do DF. Precisamos normalizar isso.abs

# Isso acontece quando o JSON tem dados aninhados, o Dataframe usando pd.METODO não consegue jogar isso num DF diretamente.

df_hosting = pd.read_json('../../data/dados_hospedagem.json')

df_hosting
```

### Transformando JSON com muitos aninhados em um DF

```python
# Normalizando dados JSON para ficar igual a um dataframe

df_hosting = pd.json_normalize(df_hosting['info_moveis'])
df_hosting
```

```python
columns = list(df_hosting.columns)

columns
```

```python
# As colunas de descrição_local para frente estavam com caracteres que não gostariamos. 
# Aqui usamos o 'explode' metódo, para remover esses caracteres da coluna 3 em diante


df_hosting = df_hosting.explode(column=columns[3:])

df_hosting
```

```python
df_hosting.reset_index(inplace=True, drop=True)

df_hosting.head()
```

```python
df_hosting.info()
```

### Transformando coluna: Máximo Hospedes

```python
# Convertendo uma coluna que antes era object para INT64 [numericos] usando .astype


df_hosting['max_hospedes'] = df_hosting['max_hospedes'].astype(np.int8)

df_hosting['max_hospedes'].dtype
```

### Transformando multiplas colunas

```python
col_numerics = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
# Convertendo multiplas colunas

df_hosting[col_numerics] = df_hosting[col_numerics].astype(np.int8)

df_hosting.dtypes
```

### Transformando coluna: Avaliação geral

```python
# Convertendo uma coluna que antes era object para FLOAT64 [float] usando .astype
df_hosting['avaliacao_geral'] = df_hosting['avaliacao_geral'].astype(np.float16)

df_hosting['avaliacao_geral'].dtype
```

```python
df_hosting.info()
```

### Transformando coluna: Preço

```python
# Remove $ e , por vazio

df_hosting['preco'] = df_hosting['preco'].apply(lambda p: p.replace('$', '').replace(',', '').strip())
```

```python
df_hosting['preco'] = df_hosting['preco'].astype(np.float64)

df_hosting['preco'].dtype
```

### Transformando multiplas colunas

```python
# .map é usado para DF

# .apply é usado para Series

# Remove todos os caracteres que não queremos e convertendo o type object para float32

df_hosting[['taxa_deposito', 'taxa_limpeza']] = df_hosting[['taxa_deposito', 'taxa_limpeza']].map(lambda p: p.replace('$', '').replace(',', '').strip())


df_hosting[['taxa_deposito', 'taxa_limpeza']] = df_hosting[['taxa_deposito', 'taxa_limpeza']].astype(np.float32)
df_hosting[['taxa_deposito', 'taxa_limpeza']].dtypes
```

### Utilizando metódos de string

```python
# Deixa dados minusculos

df_hosting['descricao_local'] = df_hosting['descricao_local'].str.lower()

df_hosting['descricao_local'].head()
```

```python
df_hosting['descricao_local'][3169]
```

```python
# Limpeza de dados da coluna usando regex

df_hosting['descricao_local'] = df_hosting['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
# Removendo hifens fora de palavras compostas

df_hosting['descricao_local'] = df_hosting['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
df_hosting['descricao_local']
```

```python
df_hosting['descricao_local'] = df_hosting['descricao_local'].str.split()

df_hosting.head()
```

```python
df_hosting['comodidades'] = df_hosting['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
df_hosting['comodidades'] = df_hosting['comodidades'].str.split(',')
```

```python
df_hosting['comodidades']
```

```python
df_hosting['descricao_vizinhanca'] = df_hosting['descricao_vizinhanca'].astype('string')
```

```python
# Transformamos o texto em letras minúsculas
df_hosting['descricao_vizinhanca'] = df_hosting['descricao_vizinhanca'].str.lower()
```

```python
# Substituimos os caractere especiais
df_hosting['descricao_vizinhanca'] = df_hosting['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
df_hosting['descricao_vizinhanca'] = df_hosting['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

### Trabalhando com Datetime

```python
import requests
from http import HTTPStatus
```

```python
r = requests.get('https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/moveis_disponiveis.json')

if r.status_code == HTTPStatus.OK:
  df_data = pd.DataFrame(r.json())

df_data.head()
```

```python
df_data.dtypes
```

```python
# Transformando tipo da coluna para Datetime

df_data['data'] = pd.to_datetime(df_data['data'])
```

```python
df_data.dtypes
```

```python
df_data.head()
```

```python
df_data['data'].dt.strftime('%Y-%m')
```

```python
subset = df_data.groupby(df_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
```

```python
subset.to_frame()
```
